Exploratory Data Analysis of COVID-19 Clinical Trials

Objective

The objective of this project is to perform Exploratory Data Analysis (EDA) on COVID-19 clinical trials data to understand trial characteristics, phases, participant demographics, geographical distribution, and trends over time.

Dataset Description

The dataset is sourced from ClinicalTrials.gov and contains information about COVID-19 related clinical studies conducted worldwide. Each row represents one clinical trial.

Tools Used


*   PYTHON
*   PANDAS
*   NUMPY
*   MATPLOTLIB
*   SEABORN














```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Loading Dataset
df = pd.read_csv("/content/COVID clinical trials.csv")

#First 5 rows
df.head()
```

```python
#Dataset Shape
df.shape
```

```python
#Column Names
df.columns
```

```python
#Data types and basic info
df.info()
```

```python
#Checking missing values count
df.isnull().sum()
```

```python
#Percentage of missing value
df.isnull().mean() * 100
```

```python
#Dropping columns with extremely high values
df.drop(["Results First Posted", "Study Documents"], axis=1, inplace=True)

#New shape
df.shape

```

```python
#Categorical columns
categorical_cols = df.select_dtypes(include=["object"]).columns
categorical_cols
```

```python
#Filling missing values in categorical columns
for col in categorical_cols:
  df[col] = df[col].fillna("Missing")
```

```python
#Confirming
df.isnull().sum()
```

```python
# checking skewness of Enrollment
df["Enrollment"].skew()
```

```python
#Finding median of Enrollment
median_enrollment = df["Enrollment"].median()

median_enrollment
```

```python
# Filling missing Enrollemnt values using median
df["Enrollment"] = df["Enrollment"].fillna(median_enrollment)

#Checking missing values again
df.isnull().sum()
```

```python
#Total number of trials
df.shape[0]
```

```python
df["Status"].value_counts()
```

```python
# Plot status distribution
plt.figure(figsize=(10,6))
df["Status"].value_counts().plot(kind="bar")
plt.title("Distribution of Clinical Trial Status")
plt.xlabel("Status")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Insight : Status distribution

From this graph, we see that the majority of COVID-19 clinical trials are in the "Recruiting" status (2805).This indicates that a large number of studies are still actively enrolling participants.
The second most common statuses are "Completed" and "Not Yet Recruiting", suggesting that many trials have either finished or are in preparation phase.  

There is an extremely small number of studies which were terminated, suspended or approved for marketing, indicating that the vast majority of studies are continuing and not being discontinued.

```python
#counting Phase values
df["Phases"].value_counts()
```

```python
# Visualizing it
plt.figure(figsize=(10,6))
df["Phases"].value_counts().plot(kind="bar")
plt.title("Distribution of Clinical Trial Phases")
plt.xlabel("Phases")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Insight: Phase Distribution

The majority of trials fall under the "Missing" and "Not Applicable" categories,
indicating that many COVID-19 studies are observational in nature.
Among the interventional trials, Phase 2 is the most common, followed by phase 3.

This suggests that most clinical trials were in mid-stage development, focusing on evaluating effectiveness and safety.

```python
#Study type
df["Study Type"].value_counts()
```

```python
# Visualising Study type
plt.figure(figsize=(10,6))
df["Study Type"].value_counts().plot(kind="bar")
plt.title("Distribution of Study Type")
plt.xlabel("Study Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Insight: Study Type Distribution
The majority of COVID-19 clinical trials are Interventional, indicating active testing of drugs, vaccines, or treatments.
A significant number are Observational, suggesting research focused on monitoring patient outcomes and disease progression without direct intervention.
This highlights the global effort to both treat and understand COVID-19 during the pandemic.

```python
#Country Distribution

df["Locations"].head()
```

```python
#Extracting country from Location
df["Country"] = df["Locations"].apply(lambda x:x.split(",")[-1].strip())
```

```python
#Verifying New Coloumn
df["Country"].head()
```

```python
#Counting Trials by country
df["Country"].value_counts().head(10)
```

```python
# Exclude "Missing" from country counts
country_counts = df[df["Country"] != "Missing"]["Country"].value_counts().head(10)

#Visualising Country Distribution
plt.figure(figsize=(10,6))
country_counts.plot(kind="bar")
plt.title("Top 10 Countries Conducting COVID-19 Clinical Trials (Excluding Missing)")
plt.xlabel("Country")
plt.ylabel("Number of Trials")
plt.xticks(rotation=45)
plt.show()
```

Insight: Country Distribution (Excluding Missing Values)


After excluding records with missing country information, the United States emerges as the leading country in conducting COVID-19 clinical trials.
France and the United Kingdom follow as the next major contributors.
This indicates that North America and Europe played a dominant role in COVID-19 research efforts during the pandemic.
The high concentration of trials in developed nations reflects strong healthcare research infrastructure, funding availability, and rapid response capabilities.
Overall, the distribution highlights the global collaboration in addressing the COVID-19 crisis, with significant contributions from both Western and Asian countries.

```python
#Enrollment Distribution Analysis(Numeric Insight)
df["Enrollment"].describe()
```

```python
#converting above values into normal values
pd.options.display.float_format = '{:,.0f}'.format
```

```python
df["Enrollment"].describe()
```

```python
#Visualizing Enrollment Distribution by limiting to 1000
plt.figure(figsize=(8,5))
sns.histplot(df[df["Enrollment"]<1000]["Enrollment"], bins=30)
plt.title("Enrollment Distribution (Below 1000 participants)")
plt.xlabel("Enrollment")
plt.ylabel("Frequency")
plt.show()

```

Insight: Enrollment Distribution

The majority of COVID-19 clinical trials have fewer than 500 participants.
The median enrollment is 170 participants, indicating that most studies were relatively small in scale.
Although the mean enrollment is significantly higher, this is due to a few extremely large trials (with enrollment in the millions), which create strong right-skewness in the data.
Overall, the data suggests that most COVID-19 trials were small to medium-sized studies rather than large population-wide trials.

```python
# Cross-Tabulation between Status and Phases

status_phase = pd.crosstab(df["Status"],df["Phases"])

status_phase
```

```python
#Visualising status vs phase
status_phase.plot(kind="bar",stacked= True, figsize=(10,6))
plt.title("Status vs Phases Distribution")
plt.xlabel("Status")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Insight: Status vs Phase Relationship

The majority of Phase 2 and Phase 3 clinical trials are in the Recruiting stage rather than Completed.
This indicates that many mid-stage and advanced trials were still ongoing at the time of data collection.
The lower number of completed Phase 3 trials suggests that long-term evaluation studies were still in progress during the pandemic.

```python
#Time trend analysis
#Converting Start date to datetime format
df["Start Date"] = pd.to_datetime(df["Start Date"],errors = "coerce")
```

```python
#Extracting month
df["Start_month"] = df["Start Date"].dt.to_period("M")


# Trial counts per month
monthly_trials =df["Start_month"].value_counts().sort_index()
monthly_trials.head()
```

```python
#Visualising Monthly Trend
plt.figure(figsize=(10,6))
monthly_trials.plot(kind="line")
plt.title("Monthly Trend of COVID-19 Clinical Trial Starts")
plt.xlabel("Month")
plt.ylabel("Number of Trials Started")
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtering data from 2019 onwards(Bcz covid started after 2019)
filtered_df = df[df["Start Date"].dt.year >= 2019].copy()

# Extract month again
filtered_df["Start_Month"] = filtered_df["Start Date"].dt.to_period("M")

monthly_trials = filtered_df["Start_Month"].value_counts().sort_index()

plt.figure(figsize=(10,6))
monthly_trials.plot(kind="line")
plt.title("Monthly Trend of COVID-19 Clinical Trial Starts (From 2019)")
plt.xlabel("Month")
plt.ylabel("Number of Trials Started")
plt.xticks(rotation=45)
plt.show()
```

Insight: Time Trend Analysis

The number of COVID-19 clinical trials increased sharply in 2020, reflecting the urgent global response to the pandemic.
The peak during 2020–2021 indicates intense research activity aimed at developing treatments, vaccines, and understanding the disease.
After the peak, trial initiation stabilized as effective treatment protocols and vaccines became widely available.

CONCLUSION

In this project, Exploratory Data Analysis (EDA) was performed on a dataset of 5,783 COVID-19 clinical trials obtained from ClinicalTrials.gov. The objective was to understand the characteristics, distribution, and trends of clinical trials conducted during the pandemic.
The analysis revealed that:

The majority of trials were in the Recruiting stage, indicating ongoing research efforts.

Interventional studies slightly outnumbered observational studies, showing active testing of drugs, vaccines, and treatments.

Among trial phases, Phase 2 and Phase 3 were the most common, and most of these trials were still recruiting rather than completed.

The United States conducted the highest number of trials, followed by France and the United Kingdom, highlighting strong research infrastructure in developed countries.

Most clinical trials had small to medium enrollment sizes, with a median of 170 participants. However, a few extremely large trials caused strong right-skewness in the data.

A sharp rise in trial initiation was observed in 2020, reflecting the rapid global scientific response to the COVID-19 pandemic.
